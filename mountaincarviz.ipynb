```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_track = np.linspace(-1.2, 0.6) # (min_position, max_position) from implementation
x_delta = 0.3
x_bottom =  - (np.pi / 2) / 3
x_heuristic = np.linspace(x_bottom - x_delta, x_bottom + x_delta)
def y(x):
    # y is the height of the track, see _height() in the implemenatation:
    # https://github.com/openai/gym/blob/b4f40f5ab621cafa778af393f540e51588fe2b8d/gym/envs/classic_control/continuous_mountain_car.py#L172
    return np.sin(3 * x) * 0.45 + 0.55 

y_track = y(x_track)
y_heurisitc = y(x_heuristic)

plt.plot(x_track, y_track, "b")
plt.plot(x_heuristic, y_heurisitc, "r")
```
